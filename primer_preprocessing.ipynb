```python
from collections import Counter
import itertools
import numpy as np
```

```python
def get_kmers(seq, k=2):
    seq = seq.upper()
    return [seq[i:i+k] for i in range(len(seq)-k+1)]
```

```python
bases = ["A", "C", "G", "T"]
all_2mers = ["".join(p) for p in itertools.product(bases, repeat=2)]
print("All possible 2-mers:", all_2mers)
```

```python
def kmer_vector(seq):
    kmers = get_kmers(seq)
    counts = Counter(kmers)
    return [counts[k] for k in all_2mers]
```

```python
X_fwd = df["F_primer"].apply(kmer_vector)
X_rev = df["R_primer"].apply(kmer_vector)
```
