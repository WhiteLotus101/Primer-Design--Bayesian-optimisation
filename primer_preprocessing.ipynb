```python
from itertools import product
from collections import Counter
import pandas as pd
```

```python
def kmer_frequency(sequence, k=2):
    """
    Returns normalized k-mer frequencies for a DNA sequence.
    """
    sequence = sequence.upper()

    kmers = [''.join(p) for p in product('ATCG', repeat=k)]

    counts = Counter(sequence[i:i+k] for i in range(len(sequence)-k+1))

    total = sum(counts.values())

    if total == 0:
        return {kmer: 0 for kmer in kmers}

    return {kmer: counts.get(kmer, 0)/total for kmer in kmers}
```

```python
import pandas as pd
import numpy as np
from itertools import product
from collections import Counter
from sklearn.model_selection import LeaveOneOut, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import ConstantKernel, RBF, WhiteKernel
from sklearn.metrics import r2_score, mean_squared_error
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("ImageJ gel analysis results_YI.xlsx", sheet_name="sheet 1", header=1)
df2 = pd.read_excel("2.xlsx", sheet_name="sheet 2", header=1)

# Normalize whitespace in headers (both files have inconsistent spacing)
df.columns = df.columns.str.replace(r"\s+", " ", regex=True).str.strip()
df2.columns = df2.columns.str.replace(r"\s+", " ", regex=True).str.strip()
df = df.rename(columns={"Log value of initial copy number of dna": "log value of copy"})

important_columns = [
    "Area",
    "Forward Primer Sequence (5'→ 3')",
    "Reverse Primer Sequence (5'→ 3')",
    "Initial copy number of DNA",
    "log value of copy",
    "forward length",
    "reverse length",
    "Run Time (minutes)",
    "Normalization"
]

df["Area"] = pd.to_numeric(df["Area"], errors="coerce")
df2["Area"] = pd.to_numeric(df2["Area"], errors="coerce")

df = df.dropna(subset=important_columns).reset_index(drop=True)
df2 = df2.dropna(subset=important_columns).reset_index(drop=True)

print(f"df (sheet 1):  {df.shape[0]} usable rows")
print(f"df2 (sheet 2): {df2.shape[0]} usable rows")
```

```python
def gc_content(seq):
    seq = seq.upper()
    return (seq.count("G") + seq.count("C")) / len(seq)

def prep(d):
    d = d.copy()
    d["F_gc"] = d["Forward Primer Sequence (5'→ 3')"].apply(gc_content)
    d["R_gc"] = d["Reverse Primer Sequence (5'→ 3')"].apply(gc_content)
    return d

combined_df = pd.concat([prep(df), prep(df2)], ignore_index=True)

print(f"combined_df: {combined_df.shape[0]} rows")
```

```python
n_outliers = (combined_df["Normalization"] > 100).sum()
print(f"Rows with Normalization > 100 before fix: {n_outliers}")

combined_df.loc[combined_df["Normalization"] > 100, "Normalization"] /= 10

print(f"Normalization range after fix: {combined_df['Normalization'].min():.2f} - {combined_df['Normalization'].max():.2f}")
```

```python
feature_cols = [
    "log value of copy",
    "forward length",
    "reverse length",
    "Run Time (minutes)",
    "F_gc",
    "R_gc"
]

X = combined_df[feature_cols]
y = combined_df["Normalization"]

print(f"X shape: {X.shape}  (was 23 rows x 37 features before these fixes)")
```

```python
feature_cols = [
    "log value of copy",
    "forward length",
    "reverse length",
    "Run Time (minutes)",
    "F_gc",
    "R_gc"
]

X = combined_df[feature_cols]
y = combined_df["Normalization"]

print(f"X shape: {X.shape}  (was 23 rows x 37 features before these fixes)")
```

```python
def loo_cv_r2(X, y, kernel_factory):
    X_arr = X.values
    y_arr = y.values
    loo = LeaveOneOut()
    preds = np.zeros(len(y_arr))

    for train_idx, test_idx in loo.split(X_arr):
        scaler = StandardScaler()
        X_train_s = scaler.fit_transform(X_arr[train_idx])
        X_test_s = scaler.transform(X_arr[test_idx])

        gpr = GaussianProcessRegressor(
            kernel=kernel_factory(),
            normalize_y=True,
            random_state=42,
            n_restarts_optimizer=8
        )
        gpr.fit(X_train_s, y_arr[train_idx])
        preds[test_idx] = gpr.predict(X_test_s)

    return preds, r2_score(y_arr, preds)

kernel_factory = lambda: (
    ConstantKernel(1.0) * RBF(length_scale=1.0, length_scale_bounds=(1e-2, 1e3))
    + WhiteKernel(noise_level=1.0, noise_level_bounds=(1e-5, 10))
)

loo_preds, loo_r2 = loo_cv_r2(X, y, kernel_factory)
loo_rmse = np.sqrt(mean_squared_error(y, loo_preds))

print(f"Leave-one-out R²:   {loo_r2:.3f}")
print(f"Leave-one-out RMSE: {loo_rmse:.2f}")
print()
print("Note: with only 23 samples, R² near 0 means the model performs about as well as")
print("predicting the mean for every point -- this is the realistic ceiling for this")
print("amount of data, not a bug. Compare to the original buggy pipeline, which scored")
print("R² = -608 (scale bug) or R² = -8.6 (scale bug fixed, but 37 features / 18 train rows).")
```

```python
plt.figure(figsize=(6, 6))
plt.scatter(y, loo_preds, alpha=0.8)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--', label="Perfect prediction")
plt.xlabel("Actual Normalization")
plt.ylabel("Predicted Normalization (LOO-CV)")
plt.title(f"Gaussian Process Regression -- LOO-CV (R2 = {loo_r2:.3f})")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
final_scaler = StandardScaler()
X_all_s = final_scaler.fit_transform(X)

final_gpr = GaussianProcessRegressor(
    kernel=kernel_factory(),
    normalize_y=True,
    random_state=42,
    n_restarts_optimizer=8
)
final_gpr.fit(X_all_s, y)

print("Fitted kernel:", final_gpr.kernel_)
print()
print("This is the model to use for new predictions. Its LOO-CV R2 above is the honest")
print("estimate of how well it will generalize -- use that number, not a single train/test split.")
```
